```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)
sine = np.sin(x)
Cosin = np.cos(x)
plt.plot(x,sine)
plt.plot(x,Cosin)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)
sine = np.sin(x)
Cosin = np.cos(x)
plt.subplot(1,2,1)
plt.plot(x,sine)
plt.subplot(1,2,2)
plt.plot(x,Cosin)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1,10,100)
log = np.log(x)

plt.plot(x,log)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)
sigmoid1 = 1/(1+np.exp(-x))
sigmoid2 = 1/(1+np.exp(-x+1))
sigmoid3 = 1/(1+np.exp(-x-1))

plt.plot(x,sigmoid1)
plt.plot(x,sigmoid2)
plt.plot(x,sigmoid3)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

zetta = 0.1
r = np.linspace(0,10,100)
DAF = 1/np.sqrt((1-r**2)**2+(2*zetta*r)**2)
plt.plot(r,DAF)


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

zetta = [0.1, 0.3, 0.5, 1.5, 2]
r = np.linspace(0,10,100)
for i in zetta:
    DAF = 1/np.sqrt((1-r**2)**2+(2*i*r)**2)
    plt.plot(r,DAF)

plt.show()
```
